```python
def carga(nombre):
    archivo=open(nombre,'r')
    return archivo.readlines()

def graf(galaxia):
    x=[]
    y=[]
    for star in galaxia:
        s=star.split()
        x.append(float(s[1]))
        y.append(float(s[2]))
    
    plot(x[0],y[0],'r*', markersize=20)
    scatter(x,y,marker='*')
```

```python
estado0=carga('condiciones_iniciales.txt')

graf(estado0)
```

```python
estado1=carga('condiciones_primer_gigayear.txt')
graf(estado1)
```
