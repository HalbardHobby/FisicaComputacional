```python
def carga(nombre):
    archivo=open(nombre,'r')
    return archivo.readlines()

def graf(galaxia):
    x=[]
    y=[]
    for star in galaxia:
        s=star.split()
        x.append(float(s[1]))
        y.append(float(s[2]))
    return (x,y)
```

```python
loc='../out/'
```

```python
estado0=carga(loc+'condiciones_iniciales.txt')

mapa=graf(estado0)
plt.figure(figsize=(5,5), dpi=100)
plot(mapa[0],mapa[1], 'b*')
plot(mapa[0][0],mapa[1][0],'g*', markersize=20)
```

```python
fig, axes = plt.subplots(1, 5, figsize=(30, 5), dpi=100)

for i in range(1,6):
    actual=carga(loc+'estado'+str(i)+'.txt')
    mapa=graf(actual)
    axes[i-1].plot(mapa[0],mapa[1], 'b*')
    axes[i-1].plot(mapa[0][0],mapa[1][0],'g*', markersize=20)
```
