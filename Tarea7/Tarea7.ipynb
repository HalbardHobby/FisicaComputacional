```python
x_max=2 #configurando las cosas obvias
t=0.15
n_u=200
n_t=200

c=1 #y las no tan obvias
dx=float(x_max)/n_u
dt=t/n_t

beta=c/(dx/dt)
print beta #al final se mira si si sirve el valor
```

```python
#cuadrando el origen del universo
x=linspace(0,x_max*pi,200)
u=3*sin(x)
u[0]=u[-1]=0

#y tomando la foto pal face *click*
plot(x,u)
```

```python
def comp_beta(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=(u[(i+1)%len(u1)]**2)-(u[i-1]**2)
        u1[i]=u1[i]*(beta/4)
    return u1

def comp_beta_sq(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=u1[i]+u[(i+1)%len(u1)]
        u1[i]=u1[i]*((u[(i+1)%len(u1)]**2)-(u[i]**2))
        u1[i]=u[i]*((beta**2)/8)
    return u1

def lax_wendroff(u):
    u1=zeros(len(u))
    
    c_bta=comp_beta(u)
    c_bta_sq=comp_beta_sq(u)
    for i in range(len(u1)):
        u1[i]=u[i]-c_bta[i]+c_bta_sq[i]
    return u1   
```

```python
u_fin=u #Se que hay mejores maneras pero no me acuerdo.
u_tot=[]
x_tot=[]
t_tot=[]

plt.figure(figsize=(11,7), dpi=100)
for t in range(n_t):
    u_fin=lax_wendroff(u_fin)
    u_tot.extend(u_fin)
    x_tot.extend(x)
    plot(x,u_fin)
    for _ in range(n_u):#Se que esto es muy poco elegante...
        t_tot.append(t*dt)#pero ya que
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(11,7), dpi=100)
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x_tot, t_tot, u_tot, rstride=10000000, cstride=10000000)
```
