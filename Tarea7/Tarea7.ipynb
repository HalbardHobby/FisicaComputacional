```python
# inicialización de las constantes del problema, sirve para ambos puntos.
x_max=2 #configurando las cosas obvias
t=0.15
n_u=200
n_t=800

c=1 #y las no tan obvias
dx=float(x_max)/n_u
dt=t/n_t

beta=c/(dx/dt)
print beta #al final se mira si si sirve el valor
```

# Condiciones iniciales, el viaje más general del problema

Se crea un array de x vacio y la velocidad. Después de eso, solo se debe hacer la gráfica de la velocidad 

```python
#cuadrando el origen del universo
x=linspace(0,x_max*pi,200)
u=3*sin(x)
u[0]=u[-1]=0

#y tomando la foto pal face *click*
plot(x,u)
xlim(0,2*pi)
```

# El método tradicional
Con el método tradicional se resuelve la ecuación diferencial parcial del despeje dado por la ecuación tres del enunciado (u^j+1,i= u^j-1,i - beta[((u^j+1)^2 -(u^j-1)^2)/2]. La ecuación se plantea por pasos, y luego se junta

```python
# definición de ecuación parcial parcial con el método 1
def trivial(u,u1):
    u2=comp(u1)
    u_r=zeros(len(u))
    for i in range(len(u_r)):
        u_r[i]=u[i]-u2[i]
    return u_r
    
def comp(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=((u[(i+1)%len(u)]**2)-(u[i-1]**2))/2
        u1[i]=beta*u1[i]
    return u1
```

# Las gráficas

Para cualquiera de los métodos, se deben llenar los vectores de velocidad y de distancia.
Después de llenarlos, simplemente se grafica el resultado que son las superposiciones del proceso

Output=gráfica

```python
#Definamos las gráficas 
u_afin=u_fin=u
u_tot=[]
x_tot=[]
t_tot=[]

#Método de gráfica (Este se utiliza para el método 2 [punto 8]
plt.figure(figsize=(11,7), dpi=100)
xlim(0,2*pi)

u_fin=trivial(u_fin,u_fin)
plot(x,u_fin)
u_in=u_fin


for t in range(n_t):
    u_fin=trivial(u_fin,u_fin)
    plot(x,u_fin)
    u_tot.extend(u_fin)
    x_tot.extend(x)
    u_afin=u_in
    u_in=u_fin
    for _ in range(n_u):#Se que esto es muy poco elegante...
        t_tot.append(t*dt)
#Quien diria que lloramos por culpa del beta...
#yo no >_<
#pero yo si TT_TT
```

# Lax Wendroff

Ahora se resuelve el problema con el método de Lax Wendroff

```python
def comp_beta(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=(u[(i+1)%len(u1)]**2)-(u[i-1]**2)
        u1[i]=u1[i]*(beta/4)
    return u1

def comp_beta_sq(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=u1[i]+u[(i+1)%len(u1)]
        u1[i]=u1[i]*((u[(i+1)%len(u1)]**2)-(u[i]**2))
        u1[i]=u[i]*((beta**2)/8)
    return u1

def lax_wendroff(u):
    u1=zeros(len(u))
    
    c_bta=comp_beta(u)
    c_bta_sq=comp_beta_sq(u)
    for i in range(len(u1)):
        u1[i]=u[i]-c_bta[i]+c_bta_sq[i]
    return u1   
```

# Informe de comparación de métodos

Se puede observar a primera medida que los dos métodos en forma son diferentes. Las funciones, sin embargo, poseen un comportamiento totalmente igual a la hora de describir la función, y eso tiene sentido dado que las funciones deben ser iguales porque se trata de la misma ecuación diferencial. Basado en lo anterior, se deben analizar las dificultades obtenidas en ambas y mencionar el mejor método.

Primero se puede observar un comportamiento diferente para el beta. Este comportamiento genera que las funciones sean más fáciles o difíciles de graficar dependiendo del valor de beta, sin embargo, este comportamiento no es igual para ambas funciones. Por ejemplo, la función tradicional presenta una dificultad en valores de beta pequeños, y no permitía un visión apropiada de la gráfica, hasta que se generó un ajuste a la misma. Este comportamiento no se mantuvo en la solución de Lax Wendroff, la cual posee una mayor tolerancia a los valores de beta en cualquier rango

Para complementar lo anterior, se puede observar que la función tradicional posee un mayor ruido que la de Lax Wendroff. Esto se debe directamente a la relación que posee con el beta. Adicionalmente, esto muestra una mayor tolerancia de la función a cambios bruscos dentro de la misma, y al pasar esto, la cantidad de ruido dentro de la función va a ser una cosa muy pequeña. Por tal razón, se puede observar una mayor precisión de la función de Lax Wendroff, y aunque ambas funciones permiten ver el comportamiento de la ecuación de Burgers, es más recomendable usar la mencionada
