```python
x_max=2 #configurando las cosas obvias
t=0.15
n_u=200
n_t=800

c=1 #y las no tan obvias
dx=float(x_max)/n_u
dt=t/n_t

beta=c/(dx/dt)
print beta #al final se mira si si sirve el valor
```

```python
#cuadrando el origen del universo
x=linspace(0,x_max*pi,200)
u=3*sin(x)
u[0]=u[-1]=0

#y tomando la foto pal face *click*
plot(x,u)
xlim(0,2*pi)
```

```python
# definición de ecuación parcial parcial con el método 1
def trivial(u,u1):
    u2=comp(u1)
    u_r=zeros(len(u))
    for i in range(len(u_r)):
        u_r[i]=u[i]-u2[i]
    return u_r
    
def comp(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=((u[(i+1)%len(u)]**2)-(u[i-1]**2))/2
        u1[i]=beta*u1[i]
    return u1
```

```python
#Definamos las gráficas 
u_afin=u_fin=u
u_total=[]
x_fin=[]
t_final=[]

#Método de gráfica (Este se utiliza para el método 2 [punto 8]
plt.figure(figsize=(11,7), dpi=100)
xlim(0,2*pi)

u_fin=trivial(u_fin,u_fin)
plot(x,u_fin)
u_in=u_fin


for _ in range(n_t):
    u_fin=trivial(u_fin,u_fin)
    plot(x,u_fin)
    u_afin=u_in
    u_in=u_fin
#Quien diria que lloramos por culpa del beta...
#yo no >_<
#pero yo si TT_TT
```

```python
def comp_beta(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=(u[(i+1)%len(u1)]**2)-(u[i-1]**2)
        u1[i]=u1[i]*(beta/4)
    return u1

def comp_beta_sq(u):
    u1=zeros(len(u))
    for i in range(len(u1)):
        u1[i]=u1[i]+u[(i+1)%len(u1)]
        u1[i]=u1[i]*((u[(i+1)%len(u1)]**2)-(u[i]**2))
        u1[i]=u[i]*((beta**2)/8)
    return u1

def lax_wendroff(u):
    u1=zeros(len(u))
    
    c_bta=comp_beta(u)
    c_bta_sq=comp_beta_sq(u)
    for i in range(len(u1)):
        u1[i]=u[i]-c_bta[i]+c_bta_sq[i]
    return u1   
```

```python
u_fin=u #Se que hay mejores maneras pero no me acuerdo.
u_tot=[]
x_tot=[]
t_tot=[]

plt.figure(figsize=(11,7), dpi=100)
for t in range(n_t):
    u_fin=lax_wendroff(u_fin)
    u_tot.extend(u_fin)
    x_tot.extend(x)
    plot(x,u_fin)
    for _ in range(n_u):#Se que esto es muy poco elegante...
        t_tot.append(t*dt)#pero ya que
```
