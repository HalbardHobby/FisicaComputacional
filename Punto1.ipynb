```python
#!/usr/bin/python
# -*- coding: latin-1 -*-
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy.fftpack import fft, fftfreq

"""
Programa con todos los metodos y funciones para el procesamiento de los datos del Encefalograma.
Autores: Oscar Fernando Garzon-Alejandro Espinosa
Fecha: JUeves 9 de Octubre de 2013.
"""
"""
Carga los datos desde un archivo "file"y los organiza en una matriz.
Input: filename.
Output: numpy array (MATRIZ) con todos los datos del encefalograma debidamente organizados.
"""
h=array( np.loadtxt('sampled+ma0844az_1-1+_data.txt', dtype='float', comments='#', delimiter=None, converters=None, skiprows=0, usecols=None, unpack=False, ndmin=0))
 

```

```python
columnas=24
filas = 400
transpuesta=empty([columnas,filas])
for i in range(filas):
    for j in range(columnas):
        transpuesta[j][i]=h[i][j]

```

```python
from scipy.fftpack import fft, fftfreq
y=empty(filas)
x_k=empty([columnas,filas/2])
frecuencias=empty([columnas,filas/2])
fft_x=zeros([columnas,filas])
for i in range(columnas):
    y=transpuesta[i]
    fft_x[i] = fft(y)/filas  # FFT Normalized
    half_n = np.ceil(filas/2.0)
    fft_x_half = (2.0) * fft_x[i][:half_n]
    freq = fftfreq(filas,1)
    freq_half = freq[:half_n]
    x_k[i]=(fft_x_half)
    frecuencias[i]=(freq_half)

for i in range(24):
    
    plt.plot((frecuencias[i]),(np.abs(x_k[i])**2))
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Power")
    plt.show()
    plt.close()


```

```python
for i in range(columnas):
    TEN_MAX= argsort(-x_k[i])[:10]
 
    for j in range (filas/2):
        if (j!=TEN_MAX[0]) & (j!=TEN_MAX[1]) & (j!=TEN_MAX[2]) & (j!=TEN_MAX[3]) & (j!=TEN_MAX[4]) & (j!=TEN_MAX[5]) & (j!=TEN_MAX[6]) & (j!=TEN_MAX[7]) & (j!=TEN_MAX[8]) & (j!=TEN_MAX[9]) :
            
            x_k[i][j]=0.0


```
